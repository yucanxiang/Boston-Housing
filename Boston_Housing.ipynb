```python
#define a funtion to generate plots to discover the relationship between the columns
def scatter(x, y, xlabel):
    plt.scatter(x,y)
    plt.xlabel(xlabel)
    plt.ylabel('House Price')
    plt.title('The Relationship between %s and house price' %xlabel)
    plt.yticks(range(0,60,5))
    plt.grid()
    plt.show()
```

```python
#Based on my observation, I decide to choose CRIM, INDUS, NOX, RM, DIS, RAD, TAX, PTRATIO, B and LSTAT columns of df to be 
#the features to train the model, and y will be the target variable.
```

```python
x=df[['CRIM','INDUS','NOX', 'RM','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df=pd.DataFrame(lm.coef_,x.columns,columns=['Coefficient'])
coeff_df
```

```python
#Interpreting the coefficients:
#From the table above, holding all other features fixed, a unit increase in the feature from each row, we can see how much 
#the house price will increase or decrease
```

```python
#Let's grab prediction from the test set and see how well it did
```

```python
predictions=lm.predict(x_test)
plt.scatter(y_test,predictions)
```

```python
import seaborn as sns
sns.distplot((y_test-predictions),bins=50)
```

```python
#Now we use root mean squared error to evaluate the model. It is a loss function. When it has a smaller value, the result 
#is better.
```

```python
from sklearn import metrics
import numpy as np
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```
